```python
server = ECMWFDataServer()
server.retrieve({
    "class": "yp",
    "dataset": "yopp",
    "date": "2018-01-01/to/2018-01-31",
    "expver": "1",
    "levtype": "sfc",
    "param": "172.128",
    "step": "0",
    "stream": "enfo",
    "time": "00:00:00",
    "type": "cf",
    "target": "yopp.grib",
})
```

```python
ds = xr.open_dataset('ecmfwsipn_land_sea_mask.grib', engine='pynio') # Typo error here (note)
ds.rename({'g0_lat_0':'lat','g0_lon_1':'lon','LSM_GDS0_SFC':'land_mask'}, inplace=True);
print(ds)
ds.land_mask.plot()
ds.to_netcdf('ecmwfsipn_land_sea_mask.nc')
```

```python
ds = xr.open_dataset('ukmetofficesipn_land_sea_mask.grib', engine='pynio')
ds.rename({'g0_lat_0':'lat','g0_lon_1':'lon','LSM_GDS0_SFC':'land_mask'}, inplace=True);
print(ds)
print('Ukmo land_mask lat/lon 0.5 degrees offset from model outout (unknown why). Shift it here...')
ds['lat'] = ds.lat + 0.5
ds['lon'] = ds.lon + 0.5
ds.land_mask.plot()
ds.to_netcdf('ukmetofficesipn_land_sea_mask.nc')
```

```python
ds = xr.open_dataset('yopp.grib', engine='pynio')
ds.rename({'g4_lat_1':'lat','g4_lon_2':'lon','LSM_GDS4_SFC':'land_mask'}, inplace=True);
ds = ds.isel(initial_time0_hours=0).drop(['initial_time0_hours','initial_time0_encoded','initial_time0'])
print(ds)
ds.land_mask.plot()
ds.to_netcdf('yopp_land_sea_mask.nc')
```
